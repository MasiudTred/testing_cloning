```python
from collections import defaultdict
```

```python
final_data = defaultdict(list)
final_data['city'] = []
print(final_data)
```

```python

for city,city_data in data.items():
    
    for rest_name,res_info in city_data.items():
       final_data['city'].append(city)
       final_data['restaurant_name'].append(rest_name)
       final_data['rating'].append(res_info['rating'])
       final_data['address'].append(res_info['address'])
    
    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(final_data)
df
```

```python
second_ques = defaultdict(list)
second_ques['city'] = []
second_ques['Dish_type']
print(second_ques)
```

```python
for city,city_data in data.items():
    
    for rest_name,res_info in city_data.items():
            for dish_type,dish_info in res_info['menu'].items():
                for dish_name,all_info in dish_info.items():
                    second_ques['city'].append(city)
                    second_ques['restaurant_name'].append(rest_name)
                    second_ques['rating'].append(res_info['rating'])
                    second_ques['address'].append(res_info['address'])
                    second_ques['Dish_type'].append(dish_type)
                    second_ques['Dish_name'].append(dish_name)
                    second_ques['Price'].append(all_info['price'])
                    second_ques['Dish_Category'].append(all_info['category'])
                        
                    second_ques['Dish_rating'].append(all_info['rating'])

    
```

```python
df = pd.DataFrame(second_ques)
df
```
